```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('Mall_Customers.csv')
df
```

```python

print("Shape:", df.shape)
print("\nInfo:\n")
print(df.info())
print("\nMissing values:\n")
print(df.isnull().sum())
```

```python
df.describe()
```

```python
df.columns = df.columns.str.strip().str.replace(" ", "_")
```

```python
df
```

```python
# Selecting only the relevant numerical features
features = df[['Age', 'Annual_Income_(k$)', 'Spending_Score_(1-100)']]
features.head()

```

```python
# Standardizing the features for KMeans
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
scaled_df = pd.DataFrame(scaled_features, columns=features.columns)
```

```python
scaled_df
```

```python
# List to store Within-Cluster Sum of Squares for each k
wcss = []

# Try K values from 1 to 10
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_df)
    wcss.append(kmeans.inertia_)  # inertia_ is the WCSS

# Plot the Elbow curve
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--', color='orange')
plt.title('Elbow Method - Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python
# Calculate silhouette score for K=2 to 10
silhouette_scores = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_df)
    score = silhouette_score(scaled_df, kmeans.labels_)
    silhouette_scores.append(score)

# Plot silhouette scores
plt.figure(figsize=(8, 5))
plt.plot(range(2, 11), silhouette_scores, marker='s', linestyle='-', color='green')
plt.title('Silhouette Score for K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
```

```python
# Apply KMeans with the selected optimal number of clusters
optimal_k = 5  # Change if needed
kmeans = KMeans(n_clusters=optimal_k, random_state=42,n_init=10)
kmeans.fit(scaled_df)

# Add cluster labels to original dataset
df['Cluster'] = kmeans.labels_
```

```python
df
```

```python
# Display cluster-wise summary
df.groupby('Cluster')[['Age', 'Annual_Income_(k$)', 'Spending_Score_(1-100)']].mean().round(2)
```

```python
# Reduce dimensions to 2D for plotting
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_df)

# Add PCA results to dataset
df['PCA1'] = pca_data[:, 0]
df['PCA2'] = pca_data[:, 1]

# Plot 2D clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='PCA1', y='PCA2', hue='Cluster', palette='Set2', s=80)
plt.title('Customer Segments Visualized in 2D (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.legend(title='Cluster')
plt.show()
```

```python
# View statistics per cluster
cluster_summary = df.groupby('Cluster')[['Age', 'Annual_Income_(k$)', 'Spending_Score_(1-100)']].mean().round(1)
cluster_summary
```

```python
# Map cluster labels to human-readable customer segments
persona_map = {
    0: "Budget Buyer",
    1: "Luxury Spender",
    2: "Casual Shopper",
    3: "Youth Saver",
    4: "Window Shopper"
}

# this map can be updated based on actual cluster_summary
df['Persona'] = df['Cluster'].map(persona_map)
```

```python
df
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Persona', palette='Set3', order=df['Persona'].value_counts().index)
plt.title("Customer Persona Distribution")
plt.xlabel("Persona")
plt.ylabel("Number of Customers")
plt.xticks(rotation=15)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
